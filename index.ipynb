```python
import pandas as pd

url = "https://github.com/Nat-sf/PC2/raw/refs/heads/main/HDI%20(1).xlsx"

# Leer el archivo Excel desde el enlace
df = pd.read_excel(url)

# Mostrar las 5 primeras filas
df.head()
```

```python
# Para saltar las 7 primeras filas

df = pd.read_excel(url, skiprows=7)
df.head()
```

```python
# Mantener las primeras 194 filas, se remueven las filas no deseadas

df = df.iloc[:194]
df
```

```python
# Se remueven las columnas con data no necesaria

df = df.drop(columns=['Unnamed: 0', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14'])
df.head()
```

```python
# Se cambian los nombres de las columnas

df.columns = ["Country", "Human_Development_Index", "Life_expectency_at_birth", "Expected_years_of_schooling", "Mean_years_of_schooling", "Gross_national_income_per_capita"]
df.head()
```

```python
# Verificar el tipo de dato que tiene cada columna del DataFrame

df.info()
```

```python
# Para cambiar el tipo de dato

columnas_a_convertir=['Human_Development_Index','Life_expectency_at_birth','Expected_years_of_schooling','Mean_years_of_schooling','Gross_national_income_per_capita']

for col in columnas_a_convertir:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#Redondeo de valores

un_decimal=['Life_expectency_at_birth','Expected_years_of_schooling','Mean_years_of_schooling']
dos_decimales=['Gross_national_income_per_capita']

df[un_decimal] = df[un_decimal].round(1)
df[dos_decimales]=df[dos_decimales].round(2)

df.head()
```

```python
# Clasificar en categoras segn el HDI

# Definir los bins y las etiquetas
bins = [0, 0.549, 0.699, 0.799, 1.0]
labels = ["LOW HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "HIGH HUMAN DEVELOPMENT", "VERY HIGH HUMAN DEVELOPMENT"]

# Crear la nueva columna 'category'
df["Category"] = pd.cut(df["Human_Development_Index"], bins=bins, labels=labels)

df.head()
```
